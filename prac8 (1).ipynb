```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
dataframe=sns.load_dataset('titanic')
```

```python
dataframe.head()
```

```python
dataframe.columns
```

```python
dataframe.shape
```

```python
#plot 1: Countplot for survival based on gender

sns.countplot(x='sex',hue='survived',data=dataframe)
plt.title('Countplot for survival based on gender')
plt.legend(title="Survival rate",labels=["no","yes"])
plt.show()
```

```python
# plot2 countplot :for survival based on passenger class

sns.countplot(x='pclass',hue='survived',data=dataframe)
plt.title('Countplot for survival based on passenger class')
plt.legend(title="Survived",labels=["no","yes"])
plt.show()
```

```python
#  histogram:age distribution of passengers by survival rate
sns.histplot(x='age',hue='survived',kde=True,multiple='stack',data=dataframe)
plt.title("Age distribution|")
plt.legend(title="Survived",labels=["no","yes"])
plt.show()
```

```python
# barplot: avg fare by class
sns.barplot(x='pclass',y='fare',estimator='mean',data=dataframe)
plt.title("Avg Fare by class")
plt.show()
```

```python
# violin plot: age vs class by survival rate
```

```python
# Age vs fare scatter plot for survival
sns.scatterplot(x='age',y='fare',hue='survived',data=dataframe)
plt.show()
```

```python
#heatmap for feature correlation
numeric=dataframe.select_dtypes(include='number')
corr=numeric.corr()
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```
